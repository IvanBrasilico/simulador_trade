```python
import pandas as pd
from pandas_datareader import data as web
dfs = {}
for acao in ['CXRI11', 'HCTR11', 'MGFF11', 'BPFF11', 'CPLE6', 'PETR4', 'RBRY11']:
    dfs[acao] = web.DataReader(acao + '.SA', data_source='yahoo', start='01-01-2021')
```

```python
dfs['PETR4'].head()
```

```python
def plot_candle(df, acao):
    df = df[df.Low > 0.]
    trace1 = {'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low}
    fig = go.Figure(go.Candlestick(**trace1))
    fig.update_layout(xaxis_rangeslider_visible=False)
    fig.show()
```

```python
plot_candle(dfs['CPLE6'], 'Copel')
```

```python
plot_candle(dfs['PETR4'], 'Petrobrás')
```

```python
plot_candle(dfs['MGFF11'], 'MGFF11')
```
