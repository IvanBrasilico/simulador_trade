```python
import pandas as pd
from pandas_datareader import data as web
from ta import add_all_ta_features
import plotly.graph_objects as go

def plot_candle(df, acao):
    df = df[df.Low > 0.]
    trace1 = {'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low}
    fig = go.Figure(go.Candlestick(**trace1))
    fig.update_layout(xaxis_rangeslider_visible=False)
    fig.show()
    
```

```python
acao = 'MGFF11.SA'
df = web.DataReader(acao, data_source='yahoo', start='01-01-2020')
```

```python
df['valor_3'] = [*list(df.Close)[3:], *[0. for r in range(3)]]
df['valor_10'] = [*list(df.Close)[10:], *[0. for r in range(10)]]
```

```python
df.valor_3 = (df.valor_3 - df.Close )/ df.Close 
df.valor_10 = (df.valor_10 - df.Close )/ df.Close
```

```python
df['valorsobe'] = (df.valor_10 + df.valor_3)
```

```python
df = df[:-20]
```

```python
df['boacompra'] = df.valorsobe > 0.04
```

```python
sum(df.boacompra) / len(df.boacompra)
```

```python
plot_candle(df, 'Vale')
```

```python
for index, row in df.iterrows():
    if row.boacompra:
        print(index)
```

```python
df_ta = add_all_ta_features(
    df, open="Open", high="High", low="Low", close="Close", volume="Volume")
```

```python
df_ta = df_ta.fillna(0.)
```

```python
df_ta
```

```python


y = df.boacompra
X = df_ta.drop(columns=['High', 'Low', 'Open', 'Close', 'Volume', 'Adj Close', 'valor_3', 'valor_10', 'valorsobe', 'boacompra'])
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X, y)
```

```python
y_pred = clf.predict(X)
```

```python
sum(y == y_pred) / len(y)
```
